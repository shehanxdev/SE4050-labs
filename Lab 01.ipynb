```python
import numpy as np
import matplotlib.pyplot as plt

num_samples = 100000
scale_parameter_exp = 1.0
exp_array = np.random.exponential(scale = scale_parameter_exp, size = num_samples)

uniform_array = np.random.rand(num_samples)

mean_normal = 0
std_normal = 1
normal_array = np.random.normal(loc = mean_normal, scale = std_normal, size = num_samples)

plt.hist(exp_array, density = True, bins = 50, histtype = "step", color = "blue", label = "Exponential")
plt.hist(uniform_array, density = True, bins = 50, histtype = "step", color = "orange", label = "Uniform")
plt.hist(normal_array, density = True, bins = 50, histtype = "step", color = "green", label = "Normal")

plt.axis([-2.5, 2.5, 0, 1.1]) 
plt.legend(loc = "upper left") 
plt.title("Random distributions") 
plt.xlabel("Value") 
plt.ylabel("Density") 
plt.show() 
```

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

x_range = np.linspace(-5, 5, 100)
y_range = np.linspace(-5, 5, 100)
X, Y = np.meshgrid(x_range, y_range)

Z = X**2 + Y**2

fig = plt.figure()
ax = fig.add_subplot(111, projection = '3d')

surface = ax.plot_surface(X, Y, Z, cmap = 'viridis')

ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')

ax.set_title('Z = X**2 + Y**2')

plt.show()
```
